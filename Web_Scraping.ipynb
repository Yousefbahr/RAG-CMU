```python
from bs4 import BeautifulSoup as Soup
import re
from pypdf import PdfReader
import os
import requests
import json
from tabula import read_pdf
import pandas as pd
```

```python
def get_Text(website, find_params=None, dont_use_params= False ,**kwargs):
  """
  website: (Str) name of website
  find: (List) list of HTML tags to search for
  """
  if find_params is None:
    find_params = ["h1", "h2", "h3", "h4", "h5", "h6", "p"]

  headers = {
      "User-Agent":"Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/137.0.0.0 Safari/537.36"
    }

  page = requests.get(website, headers=headers)

  soup = Soup(page.content, 'html.parser')

  head = soup.head
  body = soup.body

  TEXT = ""

  # title of webpage
  print(head.find("title").contents[0], end="\n\n")

  TEXT += head.find("title").contents[0] + '\n'

  if dont_use_params:
    data = body.find_all(**kwargs)

  else:
    data = body.find_all(find_params ,**kwargs)

  # print("\n\n")
  # print(data)

  for tag in data:

    TEXT += "\n"

    # main titles
    if tag.name in ["h1", "h2"]:
      TEXT += "\n- "

    # sub-titles
    elif tag.name in ["h3", "h4", "h5", "h6"]:
      TEXT += "\n-- "

    # get the text content from each tag
    for child in tag.descendants:
      if isinstance(child, str):
        TEXT += child

  print(TEXT)
  return TEXT

```

```python
def save_file(name, the_text):
  file = open(name, "w")
  file.write(the_text)
```

```python
def is_number(val):
    if pd.isna(val):
        return False
    try:
        # Remove common non-numeric characters like $, commas, and spaces
        val = str(val).replace('$', '').replace(',', '').replace('\n', '').strip()
        float(val)  # Try converting to float
        return True
    except ValueError:
        return False
```

```python
reader = PdfReader(pdf_path)

for topic, pages in topics_pages.items():
              target_dir = f"{topic}"
              target_file = f""

              text = ""

              # get tables from page if any
              # page_num is incremented because of misalignment of pages with two different pdf readers
              tables = read_pdf(pdf_path, pages= [x+1 for x in pages], stream=True)
              for j, table in enumerate(tables):

                remove_col = []
                # get duplicated column names to remove them
                for z, col in enumerate(table.columns):
                  if z == 0:
                    continue

                  if len(table) > 1 and (table.loc[1, col] == "(Decrease)" or table.loc[1, col] == "Change"):
                    continue

                  if 'Unnamed' in col:
                    # unncessary columns can sometimes have values from the next column
                    # so move those values from this column to the next
                    if z < len(table.columns) - 1:
                      sec_column = table.columns[z + 1]

                      mask = table[col].apply(is_number)
                      null_values = table.loc[:, sec_column].isnull()

                      # Combined mask where both are True
                      combined_mask = mask & null_values

                      # Move values from col1 to col2 where combined_mask is True
                      table.loc[combined_mask, sec_column] = table.loc[combined_mask, col]

                    remove_col.append(col)

                # drop unnecessary columns
                table = table.drop(columns=remove_col)

                # empty table
                if len(table.columns) == 1:
                  continue

                table.to_csv(f"{topic}/{topic}_table_{j + 1}", index=False)

              for page_num in pages:
                page = reader.pages[page_num]

                page_text = page.extract_text(extraction_mode="layout")

                # the first '\n' percedes it the page number, which I don't want so i remove it
                indx = page_text.find("\n")

                text += page_text[indx + 1:]


              save_file(f"{topic}/{topic}_article", text)

```

```python
def get_Text_2(website, find_params=None, dont_use_params= False ,**kwargs):
  """
  website: (Str) name of website
  find: (List) list of HTML tags to search for
  """
  if find_params is None:
    find_params = ["h1", "h2", "h3", "h4", "h5", "h6", "p"]

  page = requests.get(website)

  soup = Soup(page.content, 'html.parser')

  head = soup.head
  body = soup.body

  TEXT = ""

  # title of webpage
  # print(head.find("title").contents[0], end="\n\n")

  TEXT += head.find("title").contents[0] + '\n'

  if dont_use_params:
    data = body.find_all(**kwargs)

  else:
    data = body.find_all(find_params ,**kwargs)

  # print(body)
  # print("\n\n")
  # print(data)

  for tag in data:

    TEXT += "\n\n"

    # main titles
    if tag.name in ["h1", "h2"]:
      TEXT += "\n- "

    # sub-titles
    elif tag.name in ["h3", "h4", "h5", "h6"]:
      TEXT += "\n-- "

    # get the text content from each tag
    for child in tag.descendants:
      if not isinstance(child, str) and child.get("class") == "eventdate":
        TEXT += '\n'

      if child.name == 'h1':
        TEXT += '\n'

      if isinstance(child, str):
        TEXT += child


  return TEXT

```
